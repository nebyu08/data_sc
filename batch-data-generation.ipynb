```python
import torch                        # allows Tensor computation with strong GPU acceleration
from transformers import pipeline   # fast way to use pre-trained models for inference
```

```python
# Import the pandas library
import pandas as pd

# load model
dolly_pipeline = pipeline(model="databricks/dolly-v2-3b",
                            torch_dtype=torch.bfloat16,
                            trust_remote_code=True,
                            device_map="auto")
# define helper function
def get_completion_dolly(input):
  system = f"""
     generate stories based on the title that is given to you and make a great story
  """
  prompt = f"#### System: {system}\n#### User: \n{input}\n\n#### Response from Dolly-v2-3b:"
  print(prompt)
  dolly_response = dolly_pipeline(prompt,
                                  max_new_tokens=500
                                  )
  return dolly_response[0]["generated_text"]

# Define a list of prompts
prompts = ["Living with anxiety feels like..", "The day I met my soulmate..", "A zombie apocalypse in 2023.."] # add more prompts as you like

# Create an empty dataframe
df = pd.DataFrame(columns=["Prompt", "Story"])

# Loop over the prompts
for prompt in prompts:
  # Get the story from the model
  story = get_completion_dolly(prompt)
  # Add the row to the dataframe
  df = pd.concat([df, pd.DataFrame({"Prompt": [prompt], "Story": [story]})], ignore_index=True)
# Print or save the dataframe
print(df) # or df.to_csv("stories.csv") # or any other operation you want to do with the dataframe
```
