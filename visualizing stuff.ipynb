```python
heart_disease=pd.read_csv("J:/#new tutorials/Complete_Machine_Learning___Data_Science_Bootcamp/B_part1/Complete Machine Learning & Data Science Bootcamp/08 - Matplotlib Plotting and Data Visualization/013 heart-disease.csv")
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10,6)
            );
```

```python
#object oriente method mixwd with pyplot method
fix,ax=plt.subplots(figsize=(10,6));
over_50.plot(kind="scatter",
                   x="age",
                   y="chol",
                   c="oldpeak",
                   ax=ax);
ax.set_xlim([45,100]);
```

```python
heart_disease.head()
```

```python
heart_disease.sex.values
```

```python
#pure object oriented method
fig,ax=plt.subplots(figsize=(10,6))
scatt=ax.scatter(x=heart_disease["age"],
                 y=heart_disease["chol"],
                 c=heart_disease["target"])
#customize dis nuts
ax.set(title="heart disease and cholestrol",
      xlabel="age",
      ylabel="cholestrol")

#legend dis nuts
ax.legend(*scatt.legend_elements(),title="target");
#adding horizontal line
ax.axhline(heart_disease["chol"].mean(),
          linestyle="--");
```

```python
over_50.head()
```

# lets do it with 2 axes

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                          figsize=(10,10))
#lets plot the cholestrol,age and targe for the first one
scatter=ax0.scatter(x=over_50["age"],
           y=over_50["chol"],
           c=over_50["target"])
#lets customize  the first axis
ax0.set(title="the heart disease and cholestrol plot",
        xlabel="age",
       ylabel="cholestrol"
       )
#lets add the legend
ax0.legend(*scatter.legend_elements(),title="target")
#add a meanline 
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

#add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"])
#add some customization
ax1.set(title="heart disease and Max Heart rate",
       xlabel="age",
       ylabel=" Max Heart rate")
#adding a legend
ax1.legend(*scatter.legend_elements(),title=True)
#add a mean 
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--")
```
