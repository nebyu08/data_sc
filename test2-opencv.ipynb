# merging and spliting of images

```python
im4=cv2.imread("/kaggle/input/image-test/New_Zealand_Lake.jpg",1)
b,g,r=cv2.split(im4)
#show the image
plt.figure(figsize=[20,10])
plt.subplot(141);plt.imshow(b,cmap="gray");plt.title("blue gray");
plt.subplot(142);plt.imshow(g,cmap="gray");plt.title("green gray");
plt.subplot(143);plt.imshow(r,cmap="gray");plt.title("red gray");
#lets merge the chennels
rearranged=cv2.merge((b,g,r))
plt.subplot(144);plt.imshow(rearranged[:,:,::-1]);plt.title("color form");
```

```python
#changing color spaces
file="/kaggle/input/image-test/New_Zealand_Lake.jpg"
f_bgr=cv2.imread(file,1)
grb2rgb_format=cv2.cvtColor(f_bgr,cv2.COLOR_BGR2RGB)#first lets import in bgr format
plt.imshow(grb2rgb_format)
```

```python
hsv_format=cv2.cvtColor(f_bgr,cv2.COLOR_BGR2HSV)
h,s,v=cv2.split(hsv_format)

#displaying
plt.figure(figsize=[20,10])
plt.subplot(141);plt.imshow(h,cmap="gray");plt.title("Hue")
plt.subplot(142);plt.imshow(s,cmap="gray");plt.title("saturation")
plt.subplot(143);plt.imshow(v,cmap="gray");plt.title("value")

plt.subplot(144);plt.imshow(f_bgr);plt.title("orginal")
```

```python
h_new=h+10
merged=cv2.merge((h_new,s,v))
img_rgb=cv2.cvtColor(merged,cv2.COLOR_HSV2RGB)
#CHANED FORM
plt.figure(figsize=[20,10])
plt.subplot(141);plt.imshow(h,cmap="gray");plt.title("Hue")
plt.subplot(142);plt.imshow(s,cmap="gray");plt.title("saturation")
plt.subplot(143);plt.imshow(v,cmap="gray");plt.title("value")

plt.subplot(144);plt.imshow(img_rgb);plt.title("modified")
```
