```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## create data for the bar plot

# data sizes
m = 30 # rows
n =  6 # columns

# generate data
data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2   #random normal distribution
```

```python
plt.boxplot(data.T)
plt.show()
```

```python
sns.boxplot(data=data,orient='v')
plt.show()
```

```python
df=pd.DataFrame(data,columns=["zero","one","two","three","four","five",])
sns.boxplot(data=df,orient='h')
plt.show()
```
